```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

#https://api.darksky.net/forecast/[key]/[latitude],[longitude]

import requests

berlin = [52.520008, 13.404954]

with open("./data/DarkSkyAPIkey.txt", "r") as f:
    key = f.readline()

r = requests.get('https://api.darksky.net/forecast/%s/%d,%d' % (key,berlin[0],berlin[1]))

print (r.status_code)
print (r.headers['content-type'])

data = r.json()
```
